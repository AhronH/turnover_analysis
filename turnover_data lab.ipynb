```python
# Import tool and dataset
import pandas as pd
data = pd.read_csv(r"D:\raw data\MFG10YearTerminationData.csv")
```

```python
data.info()
```

```python
# Change dtype of dates
data['birthdate_key'] = pd.to_datetime(data['birthdate_key'])
data['orighiredate_key'] = pd.to_datetime(data['orighiredate_key'])
data['terminationdate_key'] = pd.to_datetime(data['terminationdate_key'])
data['recorddate_key'] = pd.to_datetime(data['recorddate_key'])
data[['birthdate_key','orighiredate_key','terminationdate_key','recorddate_key']].info()
```

```python
data.describe()
```

```python
# Change type of columns
data=data.astype({'STATUS':'category','store_name':'string'})
data[['store_name','STATUS']]
```

```python
#Split job title to take the position only, delimiter ","
data.job_title.unique() #Display unique values from the column
data[['job_title.1','job_title.2']]=data['job_title'].str.split(',', expand=True) #Split the job_title into 2 columns, delimiter ","
data[['job_title.1','job_title.2']]
data['job_title.1'].unique() #Checking unique values
data=data.drop(['job_title','job_title.2'],axis=1) #Drop old and non-essential columns
```

```python
data = data.rename(columns={'job_title.1':'job_title'}) #Rename column
```

```python
data.head(5)
#Checking data attributes. Notice that terminationdate_key occurs beforwe orighiredate_key, and ID has repetitive records with ACTIVE STATUS
```

```python
#Filter all rows with termination date occur before hire date OR Active status. "|" displays OR
index_def = data[(data["terminationdate_key"]<data["orighiredate_key"]) | (data["STATUS"]=='ACTIVE')].index
data2 = data.drop(index_def).reset_index()
data2
```

```python
#Export processed data in pkl and csv files
data2.to_pickle(r"D:\turnover_data\turnover_data_clean.pkl")
data2.to_csv(r"D:\turnover_data\turnover_data_clean.csv")
```
